```python
import pyvista as pv
```

```python

grid=pv.read("wind_channel.vtk")
grid.plot(cpos='xy')
```

```python
import numpy as np
grid.cell_neighbors(1)
grid.point_cell_ids(10)
num_points = grid.n_points
num_cells_of_point = np.array([len(grid.point_cell_ids(i)) for i in range(num_points)])
boundary_points = num_cells_of_point <= 5
inner_points = np.where(num_cells_of_point > 5)
# display(boundary_points)
# display(grid.points[boundary_points])
# display(grid.points[inner_points])
# plot the boundary points
left_boundary = grid.points[:,0] == 0
right_boundary = grid.points[:,0] == 2
top_boundary = grid.points[:,1] == 0.5
bottom_boundary = grid.points[:,1] == 0
elem_boundary = boundary_points &  ~(left_boundary | right_boundary | top_boundary | bottom_boundary)
display(grid.points[left_boundary])
display(grid.points[right_boundary])
display(grid.points[top_boundary])
display(grid.points[bottom_boundary])
display(grid.points[elem_boundary])
```
